```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
conc = [98.44,96.88,93.75,87.5,75.0,50.0,25.0,12.5,6.25,3.125,1.5625,0.78125]

Absorbance= [0.310,0.310,0.309,0.308,0.304,0.296,0.288,0.285,0.283,0.282,0.281,0.280]

wavelengths  = [400, 420, 440, 460, 480, 500, 560, 580, 600, 620, 640, 660, 680, 700,720, 740, 760, 780, 800, 820, 840, 860, 880, 900]

Transmittance= [31.0, 31, 31, 31, 31, 31, 31.2, 31.9, 42.5, 59.2, 75.6, 86.3, 95.5, 95.2, 80.1, 59.2, 50.4, 64.5, 82.7, 93.9, 96, 92.1, 83, 70.4]
```

```python
# Fit a line (degree 1 polynomial)
coeffs = np.polyfit(conc, Absorbance, 1)
bestfit = np.poly1d(coeffs)

plt.figure(figsize=(6, 5))
plt.scatter(conc, Absorbance, color='red', marker='o', s=10, label='Absorbance')
plt.plot(conc, bestfit(conc), color='black', linestyle='-', linewidth=1, label='Best Fit Line')
plt.xlabel('Concentration (%)')
plt.ylabel('Absorbance')
plt.title('Absorbance vs Concentration')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(wavelengths, Transmittance, marker='o', label='Transmittance')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Transmittance (%)')
plt.title('Transmittance vs Wavelength')
plt.legend()
plt.grid()
plt.show()
```

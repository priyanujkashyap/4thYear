
$$
\frac{dN}{dt}=rN \longrightarrow \text{Governing Equation}
$$

```python
# Population growth and Logistic Mapping
# dN/dt = rM
# N(i+1) - N(i) = rN(i)dt
# N(i+1) = N(i)(1+rdt)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Class Code
```

```python
#Code
r= np.linspace(0, 1, 10)
dt=0.1
N=[2]
i=0
while i<len(r):
    t=[0]
    while t[-1]<10:
        N.append(N[-1]*(1+r[i]*dt))
        t.append(t[-1]+dt)
        #print(t)
    plt.plot(t,N)
    i+=1
plt.show()
    
```

```python
# Population growth for different r values
r = np.linspace(0, 1, 10)
dt = 0.1

for i in range(len(r)):
    N = [2]
    t = [0]
    while t[-1] < 10:
        N.append(N[-1] * (1 + r[i] * dt))
        t.append(t[-1] + dt)
    plt.plot(t, N, label=r[i])

plt.xlabel('Time')
plt.ylabel('Population N')
plt.legend()
plt.show()
```

```python
# Population growth for different r values
r = np.linspace(2, 4, 101)
dt = 1
x_for_r = np.zeros((100, len(r)))
print(x_for_r.shape)
for i in range(len(r)):
    t = 1
    x= [0.5]
    while t<= 900:
        x.append(x[-1]*(r[i] * dt * (1 - x[-1])))
        t+=dt
    j=0
    while t<1000:
        x.append(x[-1]*(r[i] * dt * (1 - x[-1])))
        x_for_r[j][i]=x[-1]
        t+=dt
        j+=1
    
for i in range (101):
    for j in range(100):
        if x_for_r[j][i] != 0:
            plt.plot(r[i],x_for_r[j][i],',k')


plt.xlabel('Time')
plt.ylabel('Population N')
plt.show()
```

### Algorithm: Exponential Population Growth for Different r Values

1. **Initialize Parameters**
   - Create an array `r` of 10 evenly spaced values between 0 and 1.
   - Set the time step `dt` to 0.1.

2. **For Each Growth Rate (`r[i]`):**
   - Initialize a list `N` with the initial population value (2).
   - Initialize a list `t` with the initial time value (0).
   - **While the last time value in `t` is less than 10:**
     - Compute the next population value using the formula:  
       `N_next = N[-1] * (1 + r[i] * dt)`
     - Append `N_next` to the list `N`.
     - Compute the next time value:  
       `t_next = t[-1] + dt`
     - Append `t_next` to the list `t`.
   - Plot the population `N` versus time `t` for the current `r[i]`, labeling the curve with the value of `r[i]`.

3. **Finalize Plot**
   - Label the x-axis as "Time".
   - Label the y-axis as "Population N".
   - Add a legend to distinguish different `r` values.
   - Display the plot.

### Algorithm: Logistic Population Growth for Different r Values

1. **Initialize Parameters**
   - Create an array `r` of 10 evenly spaced values between 0 and 1.
   - Set the time step `dt` to 0.1.

2. **For Each Growth Rate (`r[i]`):**

````python
# Logistic population growth for different growth rates
r = np.linspace(0, 1, 10)
dt = 0.1
for i in range(len(r)):
    t = [0]
    x= [0.5]
    while t[-1] < 10:
        x.append(x[-1]*(1 + r[i] * dt * (1 - x[-1])))
        t.append(t[-1] + dt)
    plt.plot(t, x,label=f'r={r[i]:.2f}')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
````

### Algorithm: Bifurcation Diagram for the Logistic Map

1. **Initialize Parameters**
   - Create an array `r` of 101 evenly spaced values between 2 and 4.
   - Set the time step `dt` to 1.
   - Create a 2D array `x_for_r` of zeros with shape (100, len(r)) to store population values.

2. **Iterate Over Each Growth Rate (`r[i]`):**
   - Set initial time `t = 1`.
   - Initialize a list `x` with the initial normalized population value (0.5).
   - **Transient Iterations:**  
     - While `t <= 900`:
       - Compute the next population value using the logistic map:  
         `x_next = x[-1] * (r[i] * dt * (1 - x[-1]))`
       - Append `x_next` to the list `x`.
       - Increment `t` by `dt`.
   - **Record Long-Term Values:**  
     - Set `j = 0`.
     - While `t < 1000`:
       - Compute the next population value as above.
       - Append `x_next` to the list `x`.
       - Store `x_next` in `x_for_r[j][i]`.
       - Increment `t` and `j` by `dt` and 1, respectively.

3. **Plot the Bifurcation Diagram**
   - For each value of `r[i]`:
     - For each recorded value `x_for_r[j][i]`:
       - If `x_for_r[j][i]` is not zero, plot the point `(r[i], x_for_r[j][i])` as a small black mark.
   - Label the x-axis as "Time".
   - Label the y-axis as "Population N".
   - Display the plot.

````python
# Bifurcation diagram for logistic map
r = np.linspace(2, 4, 101)
dt = 1
x_for_r = np.zeros((100, len(r)))
for i in range(len(r)):
    t = 1
    x= [0.5]
    while t<= 900:
        x.append(x[-1]*(r[i] * dt * (1 - x[-1])))
        t+=dt
    j=0
    while t<1000:
        x.append(x[-1]*(r[i] * dt * (1 - x[-1])))
        x_for_r[j][i]=x[-1]
        t+=dt
        j+=1
for i in range (len(r)):
    for j in range(100):
        if x_for_r[j][i] != 0:
            plt.plot(r[i],x_for_r[j][i],',k')
plt.xlabel('Growth Rate')
plt.ylabel('Population')
plt.show()
````

````python
r = np.linspace(2, 4, 101)
dt = 1
n_transient = 900
n_plot = 100

for ri in r:
    x = 0.5
    # Transient iterations (to skip initial behavior)
    for _ in range(n_transient):
        x = x * (ri * dt * (1 - x))
    # Collect points for plotting
    xs = []
    for _ in range(n_plot):
        x = x * (ri * dt * (1 - x))
        xs.append(x)
    plt.plot([ri]*n_plot, xs, ',k')

plt.xlabel('r')
plt.ylabel('Population N')
plt.show()
````

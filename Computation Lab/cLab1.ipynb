```python
#Prime Numbers
i=2
x=[]
while i<=100000:
    j=0
    prime=True
    while j < len(x):
        if i%x[j] == 0:
            prime = False
            break
        j += 1
    if prime:
        x.append(i)
    i += 1
print(len(x))
```

```python
#Collatz Conjecture
c=[]
num=int(input("Enter a number: "))
while num !=1:
    if num%2==0:
        num/=2
    else:
        num=3*num+1
    c.append(num)
plt.plot(c,color='black',linewidth=1)
plt.xlabel('Steps')
plt.ylabel('Value')
plt.title('Collatz Conjecture')
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
c=np.cos(x)
dx=1*10**(-6)
stp=1*10**(-7)
num=[]
while dx>0:
    z=(np.sin(x+dx)-np.sin(x))/dx
    dx=dx-stp
    num.append(z)
plt.plot(x,y,label='sin')
plt.plot(x, num[-1], label='diff', linewidth=10, alpha=0.5)
plt.plot(x, c, label='cos')
plt.xlabel('sin(x)')
plt.ylabel('Derivative of sin(x)')
plt.title('Derivative of sin(x) using First Principle Method')
plt.legend()
plt.show()
```

```python
#First Principle Method
x=np.linspace(0,2*np.pi,100)
y=np.sin(x)
c=np.cos(x)
dx=1*10**(-6)
z=(np.sin(x+dx)-np.sin(x))/dx
error=(c-z)/c
print('average error:', np.average(error))
plt.plot(x, z, label='diff',color='black')
plt.plot(x,c,label='cos',linewidth=5, alpha=0.3,color='black')
plt.xlabel('sin(x)')
plt.ylabel('Derivative of sin(x)')
plt.title('Derivative of sin(x) using First Principle Method')
plt.legend()
plt.show()
```
